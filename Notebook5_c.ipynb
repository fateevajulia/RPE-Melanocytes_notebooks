```python
import numpy as np
import pandas as pd

import cassiopeia as cas
```

```python
data = pd.read_csv('~/retina/notebook/11_11_for_cassiopea/RPE_MEL_fibroblast_ascidia_less5.cvg')
```

```python
data1=data.copy()
```

```python
data1
```

```python
data1=data1.set_index('Unnamed: 0')
```

```python
data1=data1.T
```

```python
data1
```

```python
cas_tree = cas.data.CassiopeiaTree(character_matrix=data1, 
                                   #cell_meta=description1
                                  )
```

```python
ilp_solver = cas.solver.ILPSolver(
    convergence_time_limit=500,
    maximum_potential_graph_layer_size=500,
    weighted=False,
    seed=1234,
)
ilp_solver.solve(cas_tree)
```

```python
cas.pl.plot_matplotlib(cas_tree, orient="right")
```
